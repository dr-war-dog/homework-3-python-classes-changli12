#### Create a python program with a python class

```python
%matplotlib inline
import matplotlib.pyplot as plt
import csv
import pandas as pd
import numpy as np

data={}

class Dataset:
    def __init__(self, data,filepath):
        self.data = data
        self.df = pd.DataFrame.from_csv(filepath)

    def selectdata(self,year1,year2,column):
        self.data[column] = []
        self.data["Year"]=[]
        for index, row in self.df.iterrows():
            if index.year > year1 and index.year < year2:
                data["Year"].append(index.year)
                data[column].append(row[column])

    def plot(self, x_column, y_column):
        plt.plot(data[x_column], data[y_column], label=y_column)
```

#### Queastion1ï¼šCreate a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 1880
Label both lines and show the legend



```python
filepath="data/us-marriages-divorces-1867-2014.csv"
d = Dataset(data,filepath)
d.selectdata(1867,1880,"Marriages_per_1000")
d.plot("Year","Marriages_per_1000")
d.selectdata(1867,1880,"Divorces_per_1000")
d.plot("Year","Divorces_per_1000")
plt.yticks(np.linspace(0, 10, 6))
plt.title("Marriages and Divorces per 1000",fontsize=12)
plt.xlabel('Year', fontsize=12)
plt.ylabel("Number of Marriages and Divorces per 1000", fontsize=8)
plt.legend()
plt.show()
```
